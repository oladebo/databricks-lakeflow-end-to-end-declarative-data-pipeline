```python
## Create a gold streaming table

## import required libraries
import dlt
from pyspark.sql.functions import col

# =====================================
# CREATE GOLD STREAMING TABLE
# =====================================

dlt.create_streaming_table(name = 'sales_stream')

# =====================================
# APPLY AUTO CDC (SCD TYPE 2)
# =====================================

dlt.create_auto_cdc_flow(name = 'sales_cdc',
                         source = 'sales_view',
                         target = 'sales_stream',
                         keys = ['sales_id'],
                         sequence_by = col("sale_timestamp")
                         )
```
