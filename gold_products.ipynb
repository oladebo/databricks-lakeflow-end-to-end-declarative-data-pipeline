```python
## Create a gold streaming table

## import required libraries

import dlt
from pyspark.sql.functions import col

# =====================================
# CREATE GOLD STREAMING TABLE
# =====================================

dlt.create_streaming_table(name = 'products_stream')


# =====================================
# APPLY AUTO CDC (SCD TYPE 2)
# =====================================
dlt.create_auto_cdc_flow(name = 'products_cdc',
                         source = 'products_view',
                         target = 'products_stream',
                         keys = ['product_id'],
                         sequence_by = col("last_updated")
                         )
```
