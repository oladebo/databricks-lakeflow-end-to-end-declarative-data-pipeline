```python
# =====================================
# IMPORT LIBRARIES
# =====================================

import dlt
from pyspark.sql.functions import col

# =====================================
# CREATE VIEW
# =====================================

@dlt.view(
  name="customers_view")
def customers_view():
    df = spark.readStream.table("customers")
    return df

# =====================================
# CREATE STREAMING TABLE
# =====================================

dlt.create_streaming_table(
    name="customers_transformed",
    comment="slowly changing dimensions type 2 for products",
)

# =====================================
# APPLY SCD TYPE 2 (AUTO CDC)
# =====================================
### Store all changes in a table as scd type 2 using Databricks Auto CDC

dlt.create_auto_cdc_flow(
  source="customers_view",
  target="customers_transformed",
  keys = ["customer_id"],
  sequence_by ="last_updated",
  ignore_null_updates=False,
  name="customers_transformed_cdc_flow",
  except_column_list=[],
  stored_as_scd_type="2")
```
