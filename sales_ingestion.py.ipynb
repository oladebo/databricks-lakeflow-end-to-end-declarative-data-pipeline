```python
## sales ingestion data

# import required libraries
import dlt
from pyspark.sql.functions import *

# ================================
# Define data quality rules
# ================================
rules = {
    "rule1": "sales_id IS NOT NULL"
}

# ================================
# Create streaming target table
# ================================
dlt.create_streaming_table(
    name="sales",
    comment="Unified sales data from Canada and Nigeria",
    expect_all_or_drop=rules
)

# ================================
# Append Canada sales stream
# ================================
@dlt.append_flow(target="sales")
def source1():
    return spark.readStream.table(
        "hubmart1_catalog.hubmart1_schema.canada_sales"
    )

# ================================
# Append Nigeria sales stream
# ================================
@dlt.append_flow(target="sales")
def source2():
    return spark.readStream.table(
        "hubmart1_catalog.hubmart1_schema.nigeria_sales"
    )




```
