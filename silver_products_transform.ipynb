```python
# =====================================
# IMPORT LIBRARIES
# =====================================

import dlt
from pyspark.sql.functions import col

# =====================================
# CREATE VIEW
# =====================================

@dlt.view(
  name="products_view")
def products_view():
    df = spark.readStream.table("products")
    return df

# =====================================
# CREATE STREAMING TABLE
# =====================================
dlt.create_streaming_table(
    name="product_transformed",
    comment="slowly changing dimensions type 2 for products",
)

# =====================================
# APPLY SCD TYPE 2 USING APPLY_CHANGES
# =====================================
### Store all changes in a table as scd type 2 using Databricks apply_changes without creating auto cdc flow

dlt.apply_changes(
  source="products_view",
  target="product_transformed",
  keys = ["product_id"],
  sequence_by =col("last_updated"),
  ignore_null_updates=False,
  except_column_list=[],
  stored_as_scd_type="2")
```
