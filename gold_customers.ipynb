```python
## Create a gold streaming table

## import required libraries
import dlt
from pyspark.sql.functions import col

# =====================================
# CREATE GOLD STREAMING TABLE
# =====================================

dlt.create_streaming_table(name = 'customers_stream')

# =====================================
# APPLY AUTO CDC (SCD TYPE 2)
# =====================================

dlt.create_auto_cdc_flow(name = 'customers_cdc',
                         source = 'customers_view',
                         target = 'customers_stream',
                         keys = ['customer_id'],
                         sequence_by = col("last_updated")
                         )
```
